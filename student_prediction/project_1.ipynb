```python
df.drop(columns=['ID No','Prog Code','YoG','CGPA','SGPA'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
def traintest(x,y):
  x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
  return x_train,x_test,y_train,y_test
```

```python
def year_to_prdict(year):
  if year>=2 and year<=4:
    if year==2:
      x=df.drop(columns=['CGPA200','CGPA300','CGPA400'])
      y=df['CGPA200']
      lis=["CGPA100","attendance",	"study_hours"]
    elif(year==3):
      x=df.drop(columns=['CGPA300','CGPA400'])
      y=df['CGPA300']
      lis=["CGPA100",	"CGPA200","attendance",	"study_hours"]
    else:
      x=df.drop(columns=['CGPA400'])
      y=df['CGPA400']
      lis=["CGPA100",	"CGPA200",	"CGPA300",	"attendance",	"study_hours"]

  return x,y,lis
```

```python
x,y,data=year_to_prdict(4)
```

```python
y
```

```python
from sklearn.compose import make_column_transformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
preprocessing=make_column_transformer((OneHotEncoder(sparse_output=False,handle_unknown='ignore'),['Gender']),
                           (StandardScaler(),data),remainder='drop')
```

```python
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import RidgeCV
pipe=make_pipeline(preprocessing,RidgeCV(alphas=[0.01,0.1,1,10,100]))
```

```python
x_train,x_test,y_train,y_test=traintest(x,y)
```

```python
pipe.fit(x_train,y_train)
```

```python
from sklearn import set_config
set_config(display='diagram')
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,root_mean_squared_error
r2=r2_score(y_test,y_pred)
print('r2 score : ',r2)
mae=mean_absolute_error(y_test,y_pred)
print('mean_absolute_error : ',mae)
rmse=root_mean_squared_error(y_test,y_pred)
print('root_mean_squared_error : ',rmse)
```

```python
from sklearn.model_selection import learning_curve
train_sizes,train_score,valid_score=learning_curve(pipe,x,y,train_sizes=np.linspace(0.1,1.0,10),cv=5,scoring='neg_root_mean_squared_error')
train_error=-train_score.mean(axis=1)
valid_error=-valid_score.mean(axis=1)
plt.plot(train_sizes,train_error,'r-+',linewidth=2,label='train')
plt.plot(train_sizes,valid_error,'b-',linewidth=2,label='valid')
plt.show()
```

```python
coefficients = pipe.named_steps['ridgecv'].coef_

feature_names = pipe.named_steps['columntransformer'].get_feature_names_out()

importance_df = pd.DataFrame({
    "Feature": feature_names,
    "Coefficient": coefficients
})

importance_df.sort_values(by="Coefficient", inplace=True)

plt.figure(figsize=(8,6))
plt.barh(importance_df["Feature"], importance_df["Coefficient"])
plt.title("Feature Importance")
plt.show()
```

```python
import joblib
for year in[2,3,4]:
    x,y,data=year_to_prdict(year)
    preprocessing=make_column_transformer((OneHotEncoder(sparse_output=False,handle_unknown='ignore'),['Gender']),
                           (StandardScaler(),data),remainder='drop')
    pipe=make_pipeline(preprocessing,RidgeCV(alphas=[0.01,0.1,1,10,100]))
    x_train, x_test, y_train, y_test=traintest(x,y)
    pipe.fit(x_train,y_train)
    joblib.dump(pipe,f"model_year_{year}.pkl")

print("All models trained and saved")
```
